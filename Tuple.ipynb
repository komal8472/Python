# Tuple Creation

```python
tup1=()     #empty tuple
print(tup1)
```

```python
tup2=(10,30,60)  # tuple of integer numbers
print(tup2)
```

```python
tup3=(10.77,30.66,60.89)      #tuple of float numbers
print(tup3)
```

```python
tup4=('komal','navira','sandy')   #tuple of string
tup4
```

```python
tup5=('riya','riddhi',25,(34,21),(56,78))   #nested tuple
tup5
```

```python
tup6=(100,40,50,'john')   #tuple of mixed data types
tup6
```

```python
tup7=('john',34,[78,56],{'riya','david'},(99,22))
tup7
```

```python
len(tup7)    #len of list
```

# Loop through a tuple

```python
mytup=('nine','eight','seven','six','five','three','two','one')
```

```python
print(mytup)
```

```python
for i in mytuple:
    print(i)
```

```python
for i in enumerate(mytuple):
    print(i)
```
