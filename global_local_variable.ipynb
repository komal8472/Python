# Global Variable

The global keyword is used inside a function to indicate that a variable refers to a gloabal variable,not a local one

```python
x=10   #global variable
def update_x():
    global x
    x+=5
update_x()
print(x)
```

# Globals

globals() is a built in function that returns a dictionary representing that current global symbol table.it allows you to access and modify global variables programmatically

```python
x=10    #global variable
def update_x():
    globals()['x'] += 5  #access and modify the global variable using the dictionary
update_x()
print(x)
```

```python
'''if user wants 5 value then above code is applicable but if user wants only 1 value then if you write
fib(1) then you will get 2 values thats why we need write the condition hear'''
def fib(n):
    a,b=0,1
    if n==1:
        print(a)
    else:
        print(a)
        print(b)
        for i in range(2,n):
            c=a+b
            a=b
            b=c
            for i in range(2,n):
                c=a+b
                a=b
                b=c
                print(c)
fib(2)
```

# Python Decorators

```python
def div(a,b):
    print(a/b)
div(4,2)
```

```python
def div(a,b):
    print(a/b)
div(2,4)
```

```python
def div(a,b):
    if a<b:
        a,b=b,a
    print(a/b)
div(2,4)
```

```python
def div(a,b):
    print(a/b)
def div_decorator(func):
    def inner(a,b):
        if a<b:
            a,b=b,a
        return func(a,b)
    return inner
div=div_decorator(div)

div(2,4)
```

```python
def my_decorator(func):
    def wrapper():
        print('somethimng is heppening before the fun')
        func()
        print('somethong is happening after the fun')

    return wrapper
@my_decorator
def say_hello():
    print('hello!')
say_hello()
```
