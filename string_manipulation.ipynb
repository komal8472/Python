# Manipulating Strings

# Escape characters

an escape character is created by typing '\' followed by the character you want to insert

 \n ---- new line

```python
print("hello there!\nHow are you?\nI\'m doing fine.")
```

# Raw strings

A raw strings entirely ignores all escape characters and prints any backslash that appears in the string.

Raw strings are mostly used for regular expression defination.

```python
print(r"hello there\nHow are you?\nI\'m doing well.")
```

# The is X string method

isalpha()   - returns 'True' if the string consists only of letters.

isalnum()   - returns 'True' if the string consists only of letters and numbers.

isdecimal() - returns 'True' if the string consists only of numbers.

isspace()   - returns 'True' if the string consists only of spaces,tabs and new-lines.

istitle()   - returns 'True' if the string consists only of words that begin with an uppercase letter followd by only lowercase characters.

# Join() and Split()

# Join()

The 'join()' method takes all the items in an iterable ,like list,dictionary,tuple or setand joins them into a string .can also specify a separator.

```python
''.join(['my','name','is','komal'])
```

```python
','.join(['cat','dog','cow'])
```

```python
'123'.join(['cat','dog','cow'])
```

```python
'abc'.join(['cat','dog','cow'])
```

# Split()

The 'Split()' method splits a 'string' into a 'list'.by default, it will use whitespace to separate the items,but you can also set another character of choice. 

```python
'my name is komal'.split()
```

```python
'myabcnameabcisabckomal'.split('abc')
```

```python
'my name is komal'.split('m')
```

```python
'my name is komal'.split(' ')
```

```python
'hello'.rjust(20,'*')
```

```python
'hello'.ljust(20,'-')
```

```python
'hello'.center(20,'=')
```

# The count method

counts the number of occurrences of a given character or substring in the string it is applied to.can be optinally provided and end index

```python
sentence='one sheep two sheep three sheep four'
sentence.count('sheep')
```

```python
sentence.count('e')
```

```python
sentence.count('e',6)
```

```python
sentence.count('e',7)
```

# Replace method

replace all occurances of a given substring with another substring .can be optinally provided a third argument to limit the number of replacements.return a new string

```python
text='hello world'
text.replace('world','planet')
```

```python
fruits='apple,banana,cherry,apple'
fruits.replace('apple','orange',1)
```

```python
sentence='i like apple,apple are my favourite fruit'
sentence.replace('apple','orange')
```
