# Hallow Square

```python
for i in range(5):
    for j in range(5):
        if i==0 or i==4 or j==0 or j==4:
            print('*',end='')
        else:
            print(' ',end='')
    print()
```

# Floyd's Triangle

```python
num=1
for i in range(1,6):
    for j in range(1,i+1):
        print(num,end=' ')
        num+=1
    print()
```

# Hallow Right Angle Triangle

```python
for i in range(1,6):
    for j in range(1,i+1):
        if j==1 or j==i or i==5:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

# Hallow Pyramid 

```python
for i in range (1,6):
    for j in range(5-i):
        print(' ',end=' ')
    for j in range(2 * i-1):
        if j==0 or j==2 * i-2 or i==5:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

# Hallow Diamond 

```python
n=5
for i in range (1,n+1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2 * i - 1):
        if j==0 or j==2 *i -2:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()

for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range (2 * i - 1):
        if j==0 or j==2 * i-2:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

# Hallow Diamond (Number Pattern)

```python
n=5
for i in range(1,n+1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        if j==0 or j==2*i-2:
            print(i,end=' ')
        else:
            print(' ',end=' ')
    print()

for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        if j==0 or j==2 * i-2:
            print(i,end=' ')
        else:
            print(' ',end=' ')
    print()
```

# Full Star Pyramid

```python
n=5
for i in range(1,n+1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2 * i-1):
        print('*',end=' ')
    print()
```

# Inverted Full Star Pyramid

```python
n=5
for i in range(n,0,-1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2 * i-1):
        print('*',end=' ')
    print()
```
